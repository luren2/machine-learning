### 常用操作

```python
data = pd.DataFrame({'group':['a', 'a', 'a', 'b', 'b', 'b', 'c', 'c', 'c'],
                 'data':[4, 3, 2, 1, 12, 3, 4, 5, 7]})
data
```

```python
temp = data.sort_values(by=["group","data"],ascending = [False, True]) #先按照group降序 再按照data升序
temp
```

#### 去掉重复的值

```python
data = pd.DataFrame({'k1':['one']*3+['two']*4,
                 'k2':[3, 2, 1, 3, 3, 4, 4]})
data
```

```python
data.sort_values(by=['k2'], ascending = False)
```

#### 查看重复值的位置

```python
data.duplicated() #两行完全相等
```

```python
data.drop_duplicates()  #去重
```

```python
data.drop_duplicates(subset='k1')  #去重
```

#### 合并

```python
data = pd.DataFrame({'food':['A1', 'A2', 'B1', 'B2', 'B3', 'C1', 'C2'],
                 'data':[1, 2, 3, 4, 5, 6, 7,]})
data
```

```python
food2Upper = {
    'A1':'A',
    'A2':'A',
    'B1':'B',
    'B2':'B',
    'B3':'B',
    'C1':'C',
    'C2':'C'
}
data['upper']= data['food'].map(food2Upper)
data
```

### 处理缺失值

```python
string_data = pd.Series(['aardvark', 'artichoke', np.nan,'avocado'])
string_data
```

#### 判断是否缺少

```python
string_data.isnull()
```

#### 删除缺失值

```python
string_data.dropna()
```

```python
data = pd.DataFrame([[1.,6.5,3.0],[2., np.nan, 3.],[np.nan,np.nan,np.nan],[np.nan,6.5, 3.]])

data
```

```python
data.dropna()
```

#### 可设置阈值 data.dropna(thresh=2) 保留至少有n个非NaN数据的行/列

```python
x = data.dropna(axis = 0,thresh=2)
x
```

```python
data  #原来的值不改变
```

#### 删除全部为缺失值的行（整个一行全为NAN才删除）

```python
data.dropna(how = 'all')
```

#### 删除全部为缺失值的列（整个一列全为NAN才删除）

```python
data[4] = np.nan
```

```python
data
```

```python
data.dropna(how = 'all', axis = 1)
```

```python
data1 = data
data1
```

```python
data1.iloc[0:1,0:3] = 100
```

```python
data1
```

#### 删除列

```python
data1.drop(4,axis=1,inplace = True)
```

```python
data1
```

#### 删除行

```python
data1.drop(3,axis=0,inplace = True)
data1
```

#### 填充缺失值

```python
data1.fillna({1:0.2,2:0})  #第一列填充0.2 第二列填充0 不改变原来data1
```

```python
data1
```

#### 前项填充

```python
data1.fillna(method = 'bfill', limit=1) #设置填充限制
```

```python
data1[2] = data1[2].fillna(data1.iloc[:,2].mean()) #用第二列平均值填充
```

```python
data1
```
