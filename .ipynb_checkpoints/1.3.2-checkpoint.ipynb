```python
import tensorflow as tf

import cv2

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
model1 = tf.keras.models.load_model('h5/sat3.h5')
model2 = tf.keras.models.load_model('h5/sat4.h5')
#调用模型 VGG11自己训练 VGG16迁移学习
```

```python
img = cv2.imread('SAT/1.jpg',1)
#读取图片
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
img = cv2.resize(img,(224,224))
img = img.reshape(1,224,224,3)
img = img/255
#图片预处理
```

```python
img.shape
```

```python
predict1 = model1.predict(img)
predict2 = model2.predict(img)
```

```python
predict1
```

```python
predict2
```
