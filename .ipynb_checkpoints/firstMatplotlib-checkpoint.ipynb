#### 设置中文字体和标题

```python
import numpy as np
import matplotlib.pyplot as plt

# plt.rcParams["font.family"]="SimHei" #中文黑体

# fig=plt.figure(facecolor="white") #背景颜色白色
plt.figure(figsize=(8,8),facecolor="white")  #创建画布
x = np.arange(0,10,0.5)

plt.subplot(221)
plt.plot(x)
plt.title("子标题1")

plt.subplot(222)
plt.plot(x)
plt.title("子标题2",loc="left",color="blue") #左对齐 字体颜色为蓝色

plt.subplot(223)
plt.plot(x)
myfontdict={"fontsize":12,"color":"green","rotation":20}  #旋转20° 可将参数格式放在字典中
plt.title("子标题3",fontdict=myfontdict)

plt.subplot(224)
plt.plot(x, color = 'g', marker = 'o', linestyle = '--', markersize = 5, markerfacecolor = 'r')
plt.title("子标题4",color='white',backgroundcolor="black")
plt.text(0,0,'在（0,0）处添加文字',fontsize = 15, color = 'r')
plt.grid(True, linestyle = '--', color = 'g')
plt.xlabel('x---',fontsize = 18)
plt.ylabel('y---', fontsize = 18)
plt.annotate('标注', fontsize=20, xy = (15,7.5), xytext= (7, 8), 
arrowprops = dict(facecolor = 'red', shrink = 0.05,headwidth = 10, headlength = 15))

plt.suptitle("全局标题",fontsize=20,color="red",backgroundcolor="yellow")

plt.tight_layout(rect=[0,0,1,0.9])  #自动调整子图，消除子图（标题）的重叠
plt.tight_layout()
plt.show()
```

#### 散点图


```python
# scatter(x,y,acale,color,marker,label)

# x:数据点的x坐标 (不可省略)
# y:数据点的y坐标 (不可省略)
# scale:数据点的大小 默认36
# color:数据点的颜色
# marker:数据点的样式 默认为‘o’（显示为圆点）
# label:图例文字

# text(x,y,s,fontsize,color)

# x:文字x的坐标 (不可省略)
# y:文字的y坐标 (不可省略)
# s:显示的文字 (不可省略)
# fontsize:文字的大小 默认为12
# color:文字的颜色  默认为黑色

# xlabel(x,y,s,fontsize,color)  x,y可省略
# ylabel(x,y,s,fontsize,color)


# xlim(xmin,xmax)  设置坐标范围
# ylim(ymin,ymax)
#tick_params(labelsize) 设置刻度文字的字号

import numpy as np
import matplotlib.pyplot as plt

# plt.rcParams["font.family"]="SimHei"       #中文黑体
plt.rcParams["font.sans-serif"]="SimHei"     #解决中文乱码
plt.rcParams['axes.unicode_minus']=False   #防止坐标轴负号显示错误

n=100
x1=np.random.normal(0,1,n)  #正态分布
y1=np.random.normal(0,1,n)

x2=np.random.uniform(-4,4,(1,n))  #浮点数
y2=np.random.uniform(-4,4,(1,n))

plt.scatter(x1,y1,color="red", marker="x", label="标准正态分布", s=50)
plt.scatter(x2,y2,color="blue", marker="o", label="均匀分布", s=20)
plt.legend(loc='best',fontsize=13)            #将图例显示出来 1-右上角

plt.xlim(-4,4)
plt.ylim(-4,4)

plt.xlabel("横坐标x",fontsize=14)
plt.ylabel("纵坐标y",fontsize=14)

plt.title("标准正态分布与均匀分布",fontsize=20)
# plt.text(2.5,2.5,"均 值：0\n标准差：1")

plt.show()
```

#### 折线图

```python
# plot(x,y,color,marker,label,linewidth,markersize)

# x:数据点的x坐标  默认0,1,2... (可省略)
# y:数据点的y坐标 (不可省略)
# color:数据点的颜色
# marker:数据点的样式 为‘o’显示为圆点
# label:图例文字
# linewidth:折线的宽度
# markersize:数据点的大小
# markerfacecolor 
# linestyle
# alpha 透明度
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["font.sans-serif"]="SimHei" 

n=24
y1=np.random.randint(27,37,n)
y2=np.random.randint(40,60,n)

plt.plot(y1,color="blue",label="温度",marker = 'o',linestyle = '--',
         linewidth = 2, markersize = 5, markerfacecolor = 'r', alpha = 0.6)
plt.plot(y2,color="red",label="湿度",marker = 'o', markersize = 5)
plt.legend(loc=1,fontsize=13)            #将图例显示出来 1-右上角  默认也是右上角

plt.xlim(0,23)
plt.ylim(20,70)

plt.xlabel("小时",fontsize=14)
plt.ylabel("测量值",fontsize=14)

plt.title("24小时温度湿度统计",fontsize=20)

plt.show()
```

#### 柱形图

```python
# bar(left,height,width,facecolor,edgecolor,label)
# left和height不能省略

import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["font.sans-serif"]="SimHei" 
plt.rcParams['axes.unicode_minus']=False  #防止坐标轴负号显示错误

y1=[32,25,16,30,24,45,40,33,28,17,24,20]
y2=[-23,-35,-26,-35,-45,-43,-35,-32,-23,-17,-22,-28]

plt.bar(range(len(y1)),y1,width=0.8,facecolor="green",edgecolor="white",label="统计量1")
plt.bar(range(len(y2)),y2,width=0.8,facecolor="red",edgecolor="white",label="统计量2")
plt.legend(loc=1,fontsize=10) 

# plt.xlim(0,len(y1))
# plt.ylim(-60,60)

plt.title("柱状图",fontsize=20)

plt.show()
```
