```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
plt.rcParams["font.sans-serif"]="SimHei"     #解决中文乱码
plt.rcParams['axes.unicode_minus']=False   #防止坐标轴负号显示错误
```

```python
data = pd.read_csv(r'E:\data\titanic\NBA数据.csv')
```

```python
data.head()
```

```python
standard = MinMaxScaler() #归一化
X = standard.fit_transform(data.iloc[:,1:]) #一步导出
```

```python
# X=data.iloc[:,1:]
```

```python
X[:5]
```

model.cluster_centers_：簇中心。三个簇那么有三个坐标。

model.labels_：聚类后的标签

model.inertia_：所有点到对应的簇中心的距离平方和 (越小越好)

```python
loss = []
for i in range(2,11):
    model = KMeans(n_clusters=i).fit(X)
    loss.append(model.inertia_)
plt.plot(range(2,11), loss)
plt.xlabel('K值')
plt.ylabel('loss')
plt.show()
```

```python
k = 4
model = KMeans(n_clusters=k).fit(X)
```

```python
data
```

```python
label = ['中上球队','下等球队','总决赛球队', '中下球队']

data['clusters'] = model.labels_ #类别 0,1,2,3
for i in range(k):
    print('clusters:', label[i])
    label_data = data[data['clusters'] == i].iloc[:, 0]
    data.loc[data['clusters'] == i, 'clusters'] = label[i]
    print(label_data.values)
```
