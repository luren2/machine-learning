```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
from torch import nn
import torch.nn.functional as F

plt.rcParams["font.sans-serif"]="SimHei"     #解决中文乱码
plt.rcParams['axes.unicode_minus']=False   #防止坐标轴负号显示错误
```

```python
class LogisticRegressionModel(nn.Module):
    def __init__(self):
#         super(LogisticRegressionModel, self).__init__() 
        super().__init__()
        self.linear = nn.Linear(1, 1)  #构造对象 1个输入一个输出 w b已在内部初始化
    
    def forward(self, x):
        y_pred = torch.sigmoid(self.linear(x))
        return y_pred
```

```python
net = LogisticRegressionModel()
```

```python
x_data = torch.tensor([1.0, 2.0, 3.0], dtype=torch.float32).reshape(-1, 1)
y_data = torch.tensor([0, 0, 1], dtype=torch.float32).reshape(-1, 1)
```

```python
loss = nn.BCELoss()  #二元交叉熵损失
optimizer = torch.optim.SGD(net.parameters(), lr = 0.5)
epochs = 2000
```

```python
sum_loss = []
for epoch in range(epochs):
    optimizer.zero_grad()
    l = loss(net(x_data),y_data)
    sum_loss.append(l)
    l.backward()
    optimizer.step()
    
    print('第{}轮，loss为{:.8f}'.format(epoch+1, float(l.item())))
```

```python
# plt.figure(figsize=(1,1),facecolor="white", dpi=300)  #创建画布
plt.plot(range(1,epochs+1), sum_loss, 'b--',)
plt.show()
```
