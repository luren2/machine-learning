```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
from torch import nn
from d2l import torch as d2l

# %matplotlib qt5
plt.rcParams["font.sans-serif"]="SimHei"     #解决中文乱码
plt.rcParams['axes.unicode_minus']=False   #防止坐标轴负号显示错误
```

```python
class LinearModel(nn.Module):
    def __init__(self):
#         super(LinearModel, self).__init__() 
        super().__init__()
        self.linear = nn.Linear(1, 1)  #构造对象 1个输入一个输出 w b已在内部初始化
    
    def forward(self, x):
        y_pred = self.linear(x)
        return y_pred
```

```python
net = LinearModel()
```

```python
x_data = torch.tensor([[1.0], [2.0], [3.0]])
y_data = torch.tensor([[2.0], [4.0], [6.0]])
```

```python
loss = nn.MSELoss()
```

```python
optimizer = torch.optim.SGD(net.parameters(), lr = 0.001)
```

```python
epochs = 500
```

```python
sum_loss = []
for epoch in range(epochs):
    optimizer.zero_grad()
    l = loss(net(x_data),y_data)
    sum_loss.append(l)
    l.backward()
    optimizer.step()
    
    print('第{}轮，loss为{:.8f}'.format(epoch+1, float(l)))
```

```python
temp_loss = np.array(sum_loss) + 1
plt.figure(figsize=(3,3),facecolor="white", dpi=300)  #创建画布
plt.plot(range(1,epochs+1), sum_loss, 'b--',)
plt.plot(range(1,epochs+1), temp_loss, 'r',)
plt.show()
```
